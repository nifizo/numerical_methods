```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import lambdify, symbols
```

```python
a = 1
b = 5

e = 0.06

f = lambda x: 1 / (2 + x)
```

```python
n = b - a
x = np.linspace(a, b, 1000)
y = f(x)

plt.plot(x, y)
plt.show()
```

```python
def trapezoidal(f, a, b, n, h):
    s = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        s += f(a + i * h)
    return h * s

def find_h_by_runge(f, a, b, e, p=2):
    n = b - a
    h = 1

    i = 1
    while True:
        print("h = ", h)
        I1 = trapezoidal(f, a, b, n, h)
        print("Ih ", i, " = ", I1)

        h /= 2
        n *= 2
        I2 = trapezoidal(f, a, b, n, h)
        i *= 2
        print("Ih ", i, " = ", I2)
        r = abs((I2 - I1) / (2 ** p - 1))
        print("r = ", r)
        
        if abs(r) < e:
            break
        h /= 2
    return h, n
```

```python
h, n = find_h_by_runge(f, a, b, e)
print("h = ", h)

Ih = trapezoidal(f, a, b, n, h)
print("Ih = ", Ih)
```

```python
x = symbols('x')
I = sp.integrate(f(x), (x, a, b))
I = float(I)
print(" = ", I)

error = abs(I - Ih)
print(" = ", error)
```
