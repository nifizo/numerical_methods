#### Написати програму, яка знаходить максимальне та мінімальне власне значення матриці. Для тестування взяти вашу матрицз лабораторної роботи №2

```python
import numpy as np

A = np.array([
        [12, 3, 5, 8, 9],
        [3, 15, 7, 6, 4],
        [5, 7, 20, 11, 14],
        [8, 6, 11, 25, 10],
        [9, 4, 14, 10, 30]
    ])
```

```python
def MaxEigenvalue(A, eps=1e-6, max_iterations=1000):
    n = A.shape[0]
    x = np.random.rand(n)
    x /= np.linalg.norm(x, ord=2)
    lambda0 = x @ A @ x
    print("x =", x)
    print("m =", np.argmax(x))
    print("lambda0 =", lambda0)
    print("x_norm =", np.linalg.norm(x, ord=2))
    print("e =", x / np.linalg.norm(x, ord=2))
    print("                                           ")
    for i in range(1, max_iterations + 1):
        x = A @ x
        x_norm = np.linalg.norm(x, ord=2)
        x /= x_norm
        lambda1 = x @ A @ x
        print("x =", x)
        print("lambda1 =", lambda1)
        if np.abs(lambda1 - lambda0) < eps:
            return lambda1
        lambda0 = lambda1
    raise ValueError("Method did not converge within the specified number of iterations.")

B = np.array([
    [2, 1, 0],
    [1, 2, 1],
    [0, 1, 2]
])
val = MaxEigenvalue(B, eps=1e-6)
print("Max Eigenvalue:", val)
```

```python
def CheckPositive(A):
    n = A.shape[0]
    for i in range(n):
        if np.linalg.det(A[:i, :i]) < 0:
            return False
    return True

def CheckSymmetric(A):
    return np.allclose(A, A.T)
```

```python
def MinEigenvalue(A, eps=1e-6):
    if not CheckPositive(A):
        print('Matrix is not positive')
        return None
    if not CheckSymmetric(A):
        print('Matrix is not symmetric')
        return None
    A_norm = MaxEigenvalue(A, eps=eps)
    print('A_norm =', A_norm)
    B = A_norm * np.eye(A.shape[0]) - A
    print('B =', B)
    B_max_eigenvalue = MaxEigenvalue(B, eps=eps)
    print('B_max_eigenvalue =', B_max_eigenvalue)

    return A_norm - B_max_eigenvalue

B = np.array([
    [2, -1, 0],
    [-1, 2, -1],
    [0, -1, 2]
])
val = MinEigenvalue(B, eps=0.5)
```
